```python
# Média de salários dos homens e mulheres
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sb


data = pd.read_csv('http://meusite.mackenzie.br/rogerio/TIC/Projects/glassdoordata.csv')

salario_homem = data[data.gender == 'Male'][data.department == 'Administration']['income'].mean()
print('Salario médio do homem = ',salario_homem)
salario_mulher = data[data.gender == 'Female'][data.department == 'Administration']['income'].mean()
print('Salario médio da mulher = ',salario_mulher)
print(f'Salario médio do homem no departamento de administração é {(1-(salario_mulher/salario_homem))*100 :.2f}% maior que o salário médio da mulher no mesmo setor')
data.head()



```

```python
#Com esta análise, podemos observar que os homens que trabalham no departamento de administração recebem cerca de 9.46% a mais que as mulheres sem contar os bonus.
```

```python
#Media de salarios para maiores de 45 anos
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sb


data = pd.read_csv('http://meusite.mackenzie.br/rogerio/TIC/Projects/glassdoordata.csv')

salario_maior45 = data[data.age > 45]['income'].mean()
print('Salario médio para maiores de 45 anos = ', salario_maior45)

salario_menor45 = data[data.age < 45]['income'].mean()
print('Salario médio para menores de 45 anos = ', salario_menor45)

comparacao = salario_menor45/salario_maior45
print(f'O salario para maiores de 45 anos é {((1-comparacao)*100) :.2f}% maior que aqueles que tem menos de 45 anos')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns


data = pd.read_csv('http://meusite.mackenzie.br/rogerio/TIC/Projects/glassdoordata.csv')
data.head()
salario_maior45 = data[data.age>45]
salario_menor45 = data[data.age<45]

plt.subplots(figsize=(15,6))


plt.plot(1,2,1)
sns.histplot(salario_maior45.income, alpha=0.6, label='Salário mais de 45 anos' )
sns.histplot(salario_menor45.income, alpha=0.6, label='Salário menos de 45 anos' )
plt.title('Histograma de Salário Anual', fontsize=14, weight='bold')
plt.legend()


plt.show()
```

```python
# Com esta análise gráfica conseguimos fazer a comparação de salários anuais entre quem tem mais de 45 anos e quem tem menos.
# É possível notar que os maiores salários estão concentrados nas pessoas com mais idade.
# Um exemplo que podemos ver é: No levantamento existem cerca de 40 pessoas com mais de 45 anos que recebem o salário de mais ou menos R$130.000,00 por ano, enquanto apenas 9 pessoas recebem esta quantia aproximadamente
```
